```python
from ultralytics import YOLO

# Hardcoded paths and settings
DATA_YAML = r"E:\dsai_group4_project\datasets\dsai-unified-dataset\dsai-unified.yaml"
PROJECT_DIR = r"E:\dsai_group4_project\model_training\object_detection\runs"

# Fine-tune YOLO11n on your unified dataset
model = YOLO("yolo11n.pt")  
model.train(
    data=DATA_YAML,
    epochs=50,
    imgsz=640,
    batch=16,
    project=PROJECT_DIR,
    name="dsai_unified_yolo11n_ft"
)

# Quick validation after training
model.val(data=DATA_YAML)
```
