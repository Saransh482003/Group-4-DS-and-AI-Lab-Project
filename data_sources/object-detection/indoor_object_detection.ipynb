```python
import os
import pandas as pd
import json
```

```python
class_mapping = {
    0: "door",
    1: "cabinetDoor",
    2: "refrigeratorDoor",
    3: "window",
    4: "chair",
    5: "table",
    6: "cabinet",
    7: "sofa",
    8: "openedDoor",
    9: "pole"
}
```

```python
train_dir = "E:/dsai_group4_project/datasets/indoor-objects-detection/train/labels"
val_dir = "E:/dsai_group4_project/datasets/indoor-objects-detection/valid/labels"

label_files = (
    [os.path.join(train_dir, f) for f in os.listdir(train_dir)] +
    [os.path.join(val_dir, f) for f in os.listdir(val_dir)]
)
```

```python
class_counts = {class_name: 0 for class_name in class_mapping.values()}
for i, label_file in enumerate(label_files):
    with open(label_file, 'r') as f:
        anotations = [line for line in f.read().split("\n") if line.strip() != ""]
        class_ids = set([int(line.split(" ")[0]) for line in anotations])
        for class_id in class_ids:
            class_name = class_mapping[class_id]
            class_counts[class_name] += 1
print(class_counts)
```

```python
with open(r"E:\dsai_group4_project\datasets\classes.json", "r") as f:
    model_classes = json.loads(f.read())
```

```python
class_counts = pd.Series(class_counts)
class_counts.sort_values(ascending=False, inplace=True)
```
