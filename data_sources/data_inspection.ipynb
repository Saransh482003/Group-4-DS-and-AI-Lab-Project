```python
from pathlib import Path
import json, matplotlib.pyplot as plt
from PIL import Image
from matplotlib.patches import Polygon

root = Path("F:/SUNRGBD/kv1/b3dodata/img_0065")
ann = json.load(open(root / "annotation2Dfinal" / "index.json", "r", encoding="utf-8"))
img = Image.open(root / "image" / Path(ann["fileList"][0]).name)
names = [o.get("name", f"obj{i}") for i, o in enumerate(ann.get("objects", []))]
polys = ann.get("frames", [{}])[0].get("polygon", [])
idxs = sorted({p.get("object") for p in polys if isinstance(p.get("object"), int)})
use_one_based = bool(idxs) and max(idxs) == len(names)

def cls_name(i):
    if not isinstance(i, int): return "obj?"
    j = i - 1 if use_one_based else i
    return names[j] if 0 <= j < len(names) else f"obj{i}"

print("Object id -> class:", {i: cls_name(i) for i in idxs})
fig, ax = plt.subplots(figsize=(11, 8)); ax.imshow(img); ax.axis("off")
for p in polys:
    xs, ys = p.get("x", []), p.get("y", [])
    if len(xs) < 3 or len(ys) < 3: continue
    lbl = cls_name(p.get("object")); cx, cy = sum(xs) / len(xs), sum(ys) / len(ys)
    ax.add_patch(Polygon(list(zip(xs, ys)), fill=False, edgecolor="lime", linewidth=2))
    ax.text(cx, cy, lbl, color="yellow", fontsize=9, ha="center", va="center")
plt.show()
```
